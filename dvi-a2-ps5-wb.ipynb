Counting the number of Total outliers across 'Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost'.

```python
# Define a function to count outliers based on IQR
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Count outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

# List of columns to check for outliers
columns_to_check = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

# Initialize total outlier count
total_outliers = 0

# Count outliers in each column and sum them
for col in columns_to_check:
    outlier_count = count_outliers(data, col)
    print(f"Number of outliers in {col}: {outlier_count}")
    total_outliers += outlier_count

print(f"Total number of outliers across all columns: {total_outliers}")
```

**6. Handling Outliers**

* From the above, it is clear that removing Outliers based on IQR( Inter Quartile Range) might remove more than 50% of the data.
* From the boxplots above, Capping/Flooring Outliers is not effective as the outliers are spread away from the max and min. The Outliers do not seem to fall within the 5th and 95th percentile threshold range.
* The best way to handle outliers is to use Log transformation and Standardization

```python
# Calculate skewness and range
def calculate_skewness_and_range(df, columns):
    results = {}
    for column in columns:
        skewness = df[column].skew()
        range_val = df[column].max() - df[column].min()
        results[column] = {'Skewness': skewness, 'Range': range_val}
    return results

# List of columns to analyze
columns_to_analyze = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

# Get the results
results = calculate_skewness_and_range(data, columns_to_analyze)

# Print the results
for column, stats in results.items():
    print(f"Column: {column}")
    print(f"  Skewness: {stats['Skewness']}")
    print(f"  Range: {stats['Range']}")
```

Based on the skewness and range values 

* Apply Log Transformation to Sales, Profit, and Shipping Cost as they are highly skewed with a large range and 
* Apply Standardization to Quantity and Discount as they are moderately skewed with small range.

```python
from sklearn.preprocessing import StandardScaler

# Apply Log Transformation
data['Sales'] = np.log1p(data['Sales'])
data['Profit'] = np.log1p(data['Profit'])
data['Shipping Cost'] = np.log1p(data['Shipping Cost'])

# Apply Standardization
scaler = StandardScaler()
data[['Quantity', 'Discount']] = scaler.fit_transform(data[['Quantity', 'Discount']])

data.head()
```

After applying Log transformation and Standardization, checking whether outliers are still present in the dataframe.

```python
from scipy import stats

# Define a function to count outliers based on IQR
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Count outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

# List of columns to check for outliers
log_columns = ['Sales', 'Profit', 'Shipping Cost']

# Initialize log outlier count
log_outliers = 0

# Count outliers in each column and sum them
for col in log_columns:
    outlier_count = count_outliers(data, col)
    print(f"Number of outliers in {col}: {outlier_count}")
    log_outliers += outlier_count

# Define a function to count outliers based on Z-scores
def count_outliers_zscore(df, columns, threshold=3):
    outlier_counts = {}
    for column in columns:
        z_scores = np.abs(stats.zscore(df[column]))
        outlier_indices = np.where(z_scores > threshold)[0]
        outlier_counts[column] = len(outlier_indices)
    return outlier_counts

# Columns to check for outliers
stand_columns = ['Quantity', 'Discount']

# Count outliers based on Z-scores
stand_outliers_count = count_outliers_zscore(data, stand_columns)

# Initialize stand outlier count
stand_outliers = sum(stand_outliers_count.values())

# Print outlier counts for standardized columns
for column, count in stand_outliers_count.items():
    print(f"Number of outliers in {column}: {count}")

print(f"Total number of outliers across all columns: {log_outliers + stand_outliers}")
```

As the total number of outliers are only 1303, we are handling it by removing it.

```python
# Function to find outlier indices based on IQR
def find_outlier_indices_iqr(df, columns):
    outlier_indices = set()
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        indices = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
        outlier_indices.update(indices)
    return outlier_indices

# Function to find outlier indices based on Z-scores
def find_outlier_indices_zscore(df, columns, threshold=3):
    outlier_indices = set()
    for column in columns:
        z_scores = np.abs(stats.zscore(df[column]))
        indices = np.where(z_scores > threshold)[0]
        outlier_indices.update(indices)
    return outlier_indices

# Columns to check for outliers
log_columns = ['Sales', 'Profit', 'Shipping Cost']
stand_columns = ['Quantity', 'Discount']

# Find outlier indices
outlier_indices_iqr = find_outlier_indices_iqr(data, log_columns)
outlier_indices_zscore = find_outlier_indices_zscore(data, stand_columns)

# Combine outlier indices
combined_outlier_indices = outlier_indices_iqr.union(outlier_indices_zscore)

# Remove rows with outliers
data_cleaned = data.drop(index=combined_outlier_indices)

# Verify the number of removed outliers
print(f"Total number of outliers removed: {len(combined_outlier_indices)}")
print(f"Original DataFrame shape: {data.shape}")
print(f"Cleaned DataFrame shape: {data_cleaned.shape}")
```
